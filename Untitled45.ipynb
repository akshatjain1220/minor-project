```python
from nltk.tokenize import word_tokenize
```

```python
def tokenize(t):
    if isinstance(t, str):
        tokens = word_tokenize(t)
        return tokens
```

```python
data['Description'] = data['Description'].apply(tokenize)
```

```python
from nltk.corpus import stopwords
```

```python
def filterstepwords(t):
    str = " "
    sw=set(stopwords.words('english'))
    for token in t:
        if token.lower() not in sw:
            str+=token.lower()
            str+=" "
    return str
```

```python
data['Description'] = data['Description'].apply(filterstepwords)
```

```python
from nltk.stem import PorterStemmer
```

```python
s=PorterStemmer()
```

```python
def stem_text(txt):
    tokens = word_tokenize(txt)
    s_tokens = []
    for token in tokens:
        s_token = s.stem(token)
        s_tokens.append(s_token)
    stemmed_txt = ' '.join(s_tokens)
    return stemmed_txt
```

```python
data['Description_stemmed'] = data['Description'].apply(stem_text)
```

```python
def recommendations(song_name,k=5):
    idx = track_names.index(song_name)
    similarities = cosine_similarity(sparse_matrix[idx-1], sparse_matrix).flatten()
    similar_indices = similarities.argsort()[::-1][1:k+1]
    similar_songs = []
    for j in similar_indices:
        similar_songs.append(track_names[j+1])
    return similar_songs
```

```python
song_name = 'All the Time - Don Diablo Remix'
recommended_songs = recommendations(song_name)
for idx, song_name in enumerate(recommended_songs, 1):
    print(f"{idx}. {song_name}")
```
